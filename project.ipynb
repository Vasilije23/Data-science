```python
data.isnull().sum()
```

```python
data.drop(['Market Category'], inplace=True, axis=1)
data.isnull().sum()
```

```python
data.dropna(inplace=True,axis=0)
data.isnull().sum()
```

```python
data.duplicated().sum()
```

```python
data.drop_duplicates(inplace=True, keep=False)
```

```python
data.info()
```

```python
sns.boxplot(data=data,orient='h',palette='Set2')
```

```python
data.drop(data[data['MSRP'] >= 600000].index,inplace=True)
```

```python
convert=['Make','Model','Engine Fuel Type','Transmission Type','Driven_Wheels','Vehicle Size','Vehicle Style']
data[convert] = data[convert].astype('category')
```

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()

for d in ['Make','Model','Engine Fuel Type','Transmission Type','Driven_Wheels','Vehicle Size','Vehicle Style']: data[d] = label_encoder.fit_transform(data[d])
```

```python
data.head()
```

```python
x = data[['Popularity','Year','Engine HP','Engine Cylinders','highway MPG','city mpg','Number of Doors']].values
y = data['MSRP'].values

print(x.shape)
print(y.shape)
```

```python
from sklearn.preprocessing import StandardScaler
ssx=StandardScaler()
ssy=StandardScaler()

x=ssx.fit_transform(x)
y=ssy.fit_transform(y.reshape(-1,1))
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.15,random_state = 0)

print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
from sklearn.ensemble import RandomForestRegressor

rf_model=RandomForestRegressor(n_estimators=100, n_jobs=-1, random_state=0, oob_score=True)
rf_model.fit(x_train,np.ravel(y_train))

print(rf_model.score(x_train,y_train))
print(rf_model.score(x_test,y_test))
print(1 - rf_model.oob_score_)
```

```python
y_pred = rf_model.predict(x_test)
y_pred
```

```python
plt.scatter(y_test,y_pred)
plt.title("Random forest regression model")
plt.xlabel("True")
plt.ylabel("Predicted")
```

```python
sns.distplot(y_test-y_pred)
```

```python
from sklearn import metrics

print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```python
oob = []
n = [10, 20, 50, 100, 200, 500, 750, 1000]
for i in n:
        rf_model.set_params(n_estimators=i, oob_score=True)
        rf_model.fit(x_train,np.ravel(y_train))
        oob_error = 1 - rf_model.oob_score_
        oob.append(oob_error)
        
print(oob)

plt.plot(n, oob,'r')
plt.ylabel("oob_error")
plt.xlabel("n_estimators")
```

```python
from sklearn.linear_model import LinearRegression

lr_model = LinearRegression()
lr_model.fit(x_train,y_train)

print(lr_model.score(x_train,y_train))
print(lr_model.score(x_test,y_test))
```

```python
y_pred = lr_model.predict(x_test)
y_pred
```

```python
plt.scatter(y_test,y_pred)
plt.xlabel("True")
plt.ylabel("Predicted")
```

```python
sns.distplot(y_test-y_pred)
```
